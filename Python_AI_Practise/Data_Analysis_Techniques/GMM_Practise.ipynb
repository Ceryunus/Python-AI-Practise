```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# rnek veri oluturma
np.random.seed(1)
# Standart sapma 0.5 olacak ekilde 100 veri ve 2 zellik olacak ekilde
# 3 ayr noktada 2,2  7,7, ve 3,10
X = np.vstack([np.random.normal(loc, 0.5, size=(100, 2)) for loc in [(2, 2), (7, 7), (3, 10)]])


# GMM (Gaussian Mixture Models) 
# Veri setlerini alt guruplara ayrmak iin kullanlr.
# K-Means a gre daha baarl.
# kmeleme mat ile alyr bir veri noktas birden fazla kmeye dahil olabiliyor bu da esneklik salar.
# her alt kme kendi guassian dalmna sahiptir

# Varyans ile ilikisi: 
# Gaussian dalm, verilerin ortalama etrafnda ne kadar yaygn olduunu gsterirken;
# varyans, bu yaylmn lsn belirler. 


# Gaussian Mixture Model oluturma ve eitme
# 3 kme olacak
gmm = GaussianMixture(n_components=3, random_state=1)
gmm.fit(X)
labels = gmm.predict(X)

# Kme merkezlerini
centers = gmm.means_

# Kmeleme sonularn grselletirme
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', edgecolor='k')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=100, label='Merkezler')
plt.title("Gaussian Mixture Modeli Kmeleme")
plt.xlabel('zellik 1')
plt.ylabel('zellik 2')
plt.legend()
plt.show()
```
