```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# Örnek veri oluşturma
np.random.seed(1)
# Standart sapma 0.5 olacak şekilde 100 veri ve 2 özellik olacak şekilde
# 3 ayrı noktada 2,2  7,7, ve 3,10
X = np.vstack([np.random.normal(loc, 0.5, size=(100, 2)) for loc in [(2, 2), (7, 7), (3, 10)]])


# GMM (Gaussian Mixture Models) 
# Veri setlerini alt guruplara ayırmak için kullanılır.
# K-Means a göre daha başarılı.
# kümeleme matığı ile çalışıyr bir veri noktası birden fazla kümeye dahil olabiliyor bu da esneklik sağlar.
# her alt küme kendi guassian dağılımına sahiptir

# Varyans ile ilişkisi: 
# Gaussian dağılımı, verilerin ortalama etrafında ne kadar yaygın olduğunu gösterirken;
# varyans, bu yayılımın ölçüsünü belirler. 


# Gaussian Mixture Model oluşturma ve eğitme
# 3 küme olacak
gmm = GaussianMixture(n_components=3, random_state=1)
gmm.fit(X)
labels = gmm.predict(X)

# Küme merkezlerini
centers = gmm.means_

# Kümeleme sonuçlarını görselleştirme
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', edgecolor='k')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=100, label='Merkezler')
plt.title("Gaussian Mixture Modeli Kümeleme")
plt.xlabel('Özellik 1')
plt.ylabel('Özellik 2')
plt.legend()
plt.show()
```
