```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

np.random.seed(42)
X = np.random.rand(100, 2)  # 100 veri noktası 2 özellik


#PCA (Principal Component Analysis)
#Verideki en büyük varyansı açıklayan doğrusal bileşenleri bulur.
#Bu, verinin yapısını ve en önemli özelliklerini anlamamızı sağlar.

#PCA nın özelliği kısaca verideki en büyük varyansı açıklayan yönleri bulması 
#böylece verinin ana yapısı ve önemli özelliklerini daha basit indirgenmiş veriler
#ile ifade etmesidir.

#Verideki boyut yüskesk ise kullanılır boyutu düşürmek amacıyla kullanılır
#Oluşturduğumuz veride 2 farklı özellik belirlenmiştir bunu 1 özelliğe yani boyuta indirgeyeceğiz.

# PCA modeli oluşturma ve veriye uygulama
pca = PCA(n_components=1)  # 1 bileşene indirgeniyor
X_pca = pca.fit_transform(X)

# Sonuçları görselleştirme
plt.scatter(X[:, 0], X[:, 1], label='Orijinal Veri')
plt.scatter(X_pca, np.zeros_like(X_pca), label='PCA ile İndirgenmiş Veri')
plt.xlabel('Özellik 1')
plt.ylabel('Özellik 2')
plt.title('PCA Uygulaması')
plt.legend()
plt.show()
```
