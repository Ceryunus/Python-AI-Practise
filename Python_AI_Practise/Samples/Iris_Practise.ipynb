```python
#Gaussian Naive Bayes 
# zellikler 
# Bayes teoremi ve Gaussian dalm kullanarak olaslksal snflandrma yapar
# Her zellii bamsz olarak deerlendirir. (x)
# Her snf iin olaslk hesaplar ve en yksek olasla sahip snf tahmin eder.
# Hzl ve basittir byk verilerde kullanlabilir.
# zellikler bamsz deilse ve normal bir dalm gsteriyorsa kullanlmamaldr.
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(x_iris, y_iris, test_size=0.2, random_state=12)
```

```python
model = GaussianNB()
```

```python
model.fit(Xtrain,ytrain);
```

```python
y_model = model.predict(Xtest)
```

```python
accuracy_score(ytest,y_model)
```

```python
#PCA (Principal Component Analysis)
#Verideki en byk varyans aklayan dorusal bileenleri bulur.
#Bu, verinin yapsn ve en nemli zelliklerini anlamamz salar.

#PCA nn zellii ksaca verideki en byk varyans aklayan ynleri bulmas 
#bylece verinin ana yaps ve nemli zelliklerini daha basit indirgenmi veriler
#ile ifade etmesidir.

#Verideki boyut yskesk ise kullanlr boyutu drmek amacyla kullanlr
#Iris verisinde her bir tr iin 4 farkl zellik belirlenmitir bunu 2 zellie yani boyuta indirgeyeceiz.
```

```python
from sklearn.decomposition import PCA
```

```python
model = PCA(n_components=2)
```

```python
# 4 Boyutlu hali
```

```python
x_iris.head()
```

```python
model.fit(x_iris);
```

```python
X_2D = model.transform(x_iris)
```

```python
type(X_2D)
```

```python
# 2 Boyuta indirgenmi hali
```

```python
X_2D[0:10]
```

```python
iris["PCA1"]= X_2D[:,0]
iris["PCA2"]= X_2D[:,1]
```

```python
iris
```

```python
#Sadece zellikleri vererek bir ayrm yapmasn istedik sonularda 
#"setosa" ok iyi bir ekilde snflandrabilimi.
```

```python
sns.scatterplot(x="PCA1",y="PCA2",hue="species",data=iris)
```

```python
# GMM (Gaussian Mixture Models) 
# Veri setlerini alt guruplara ayrmak iin kullanlr.
#  K-Means a gre daha baarl.
# kmeleme mat ile alyr bir veri noktas birden fazla kmeye dahil olabiliyor bu da esneklik salar.
# her alt kme kendi guassian dalmna sahiptir

# Varyans ile ilikisi: 
# Gaussian dalm, verilerin ortalama etrafnda ne kadar yaygn olduunu gsterirken;
# varyans, bu yaylmn lsn belirler. 
```

```python
from sklearn.mixture import GaussianMixture
```

```python
# 3 kme (trler iin) oluturacaz o yzden hiperparametersini 3 olarak ayarladk 
```

```python
model = GaussianMixture(n_components=3, covariance_type="full",random_state=1)
```

```python
model.fit(x_iris);
```

```python
y_gmm = model.predict(x_iris)
```

```python
y_gmm
```

```python
iris["cluster"] = y_gmm
```

```python
# trlerimizi 3 kmeye ayrdk
#grld zere e. grafikte 1. grafikte olan trn verileri de gzkmkete
#GMM zellii olarak birden fazla kmeye dahil olabiliyordu...
```

```python
sns.lmplot(x="PCA1", y="PCA2", data=iris, hue="species",
           col="cluster", fit_reg=False);
```
