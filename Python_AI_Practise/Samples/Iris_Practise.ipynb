```python
#Gaussian Naive Bayes 
# Özellikler 
# Bayes teoremi ve Gaussian dağılımı kullanarak olasılıksal sınıflandırma yapar
# Her özelliği bağımsız olarak değerlendirir. (x)
# Her sınıf için olasılık hesaplar ve en yüksek olasılığa sahip sınıfı tahmin eder.
# Hızlı ve basittir büyük verilerde kullanılabilir.
# Özellikler bağımsız değilse ve normal bir dağılım gösteriyorsa kullanılmamalıdır.
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(x_iris, y_iris, test_size=0.2, random_state=12)
```

```python
model = GaussianNB()
```

```python
model.fit(Xtrain,ytrain);
```

```python
y_model = model.predict(Xtest)
```

```python
accuracy_score(ytest,y_model)
```

```python
#PCA (Principal Component Analysis)
#Verideki en büyük varyansı açıklayan doğrusal bileşenleri bulur.
#Bu, verinin yapısını ve en önemli özelliklerini anlamamızı sağlar.

#PCA nın özelliği kısaca verideki en büyük varyansı açıklayan yönleri bulması 
#böylece verinin ana yapısı ve önemli özelliklerini daha basit indirgenmiş veriler
#ile ifade etmesidir.

#Verideki boyut yüskesk ise kullanılır boyutu düşürmek amacıyla kullanılır
#Iris verisinde her bir tür için 4 farklı özellik belirlenmiştir bunu 2 özelliğe yani boyuta indirgeyeceğiz.
```

```python
from sklearn.decomposition import PCA
```

```python
model = PCA(n_components=2)
```

```python
# 4 Boyutlu hali
```

```python
x_iris.head()
```

```python
model.fit(x_iris);
```

```python
X_2D = model.transform(x_iris)
```

```python
type(X_2D)
```

```python
# 2 Boyuta indirgenmiş hali
```

```python
X_2D[0:10]
```

```python
iris["PCA1"]= X_2D[:,0]
iris["PCA2"]= X_2D[:,1]
```

```python
iris
```

```python
#Sadece özellikleri vererek bir ayrım yapmasını istedik sonuçlarda 
#"setosa" çok iyi bir şekilde sınıflandırabilimiş.
```

```python
sns.scatterplot(x="PCA1",y="PCA2",hue="species",data=iris)
```

```python
# GMM (Gaussian Mixture Models) 
# Veri setlerini alt guruplara ayırmak için kullanılır.
#  K-Means a göre daha başarılı.
# kümeleme matığı ile çalışıyr bir veri noktası birden fazla kümeye dahil olabiliyor bu da esneklik sağlar.
# her alt küme kendi guassian dağılımına sahiptir

# Varyans ile ilişkisi: 
# Gaussian dağılımı, verilerin ortalama etrafında ne kadar yaygın olduğunu gösterirken;
# varyans, bu yayılımın ölçüsünü belirler. 
```

```python
from sklearn.mixture import GaussianMixture
```

```python
# 3 küme (türler için) oluşturacağız o yüzden hiperparametersini 3 olarak ayarladık 
```

```python
model = GaussianMixture(n_components=3, covariance_type="full",random_state=1)
```

```python
model.fit(x_iris);
```

```python
y_gmm = model.predict(x_iris)
```

```python
y_gmm
```

```python
iris["cluster"] = y_gmm
```

```python
# türlerimizi 3 kümeye ayırdık
#görüldüğü üzere e. grafikte 1. grafikte olan türün verileri de gözükmkete
#GMM özelliği olarak birden fazla kümeye dahil olabiliyordu...
```

```python
sns.lmplot(x="PCA1", y="PCA2", data=iris, hue="species",
           col="cluster", fit_reg=False);
```
